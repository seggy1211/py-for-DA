MovieLens 1M Data Set

```python
import pandas as pd
unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('./users.dat', sep='::', header=None, names=unames)
users.head()
```

```python
rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('./ratings.dat', sep='::', header = None, names=rnames)
ratings.head()
```

```python
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('./movies.dat', sep='::', header=None, names=mnames)
movies.head()
```

```python
test = pd.merge(ratings, users) # finds the col with the same name and merges the frames using it
test
```

```python
data = pd.merge(pd.merge(ratings, users), movies)
data.head()
```

```python
data.shape # get the number of columns and rows
specific_film = data[data['title'] == "One Flew Over the Cuckoo's Nest (1975)"] # find with the exact title 
specific_film.head()
s_word = data[data['title'].str.contains("Zero Kelvin", case=False)] # check for all matches
s_word.head()
```

```python
mean_rating = data.pivot_table(index='title', columns='gender', values='rating') # new syntax for pivot_table automatically, counts mean, devides if different columns='gender'
mean_rating.head()
```

```python
#Filtering 
ratings_by_title = data.groupby('title').size() # groupby gathers movies the same 'title' to works needs some funciton here it is size()
#ratings_by_title = data['title'].value_counts() ---> gives the same thing 
ratings_by_title.head()
active_titles = ratings_by_title.index[ratings_by_title >=250].sort_values(ascending=True) # index because index is a film name here 
active_titles[0:10]
```

```python
mean_rating = mean_rating.loc[active_titles]
mean_rating
```

```python
top_female_ratings = mean_rating.sort_values(by='F', ascending=False)
top_female_ratings.head()
```

```python
mean_rating['diff'] = mean_rating['M'] - mean_rating['F'] # adding new col
mean_rating.head()
```

```python
sorted_by_diff = mean_rating.sort_values(by='diff') # prefered more by women or men
sorted_by_diff.head()
```

```python
rating_std_by_title = data.groupby('title')['rating'].std() # combines all enteries for the film and counts std
rating_std_by_title.head()
```

```python
rating_std_by_title = rating_std_by_title.loc[active_titles]
rating_std_by_title.sort_values(ascending=True).head()
```
